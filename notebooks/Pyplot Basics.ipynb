```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot([1,2,3],[10,20,30],'ro') # If you provide a single list or array to the
                       #  plot() command, matplotlib assumes it is a 
                       # sequence of y values, and automatically 
                       # generates the x values for you.
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.show()
```

```python
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
plt.plot(range(1,5), 'ro')
plt.plot(range(2,20,2), 'p--', linewidth=10.0)
plt.plot(range(2,40,5), 'g^--')
plt.show()
```
